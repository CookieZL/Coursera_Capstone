```python
# Scrape table from Wikipedia  
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df.head()
```

```python
# Delete row from dataframe if the column Borough contains "Not assigned"
indexNames = df[df['Borough'] == 'Not assigned' ].index
df.drop(indexNames , inplace=True)
df
```

```python
df.loc[df['Neighbourhood'] == 'Not assigned', ['Borough']] = 'Neighbourhood'
df
raw_data_selected = df
```

```python
geospatial_url = "https://cocl.us/Geospatial_data"
geospatial_data = pd.read_csv(geospatial_url)
geospatial_data.head()
```

```python
geospatial_data.columns
```

```python
merged_data = pd.merge(raw_data_selected, geospatial_data, on='Postal Code')
merged_data.head()
```

```python
merged_data['Coordinates'] = list(zip(merged_data['Latitude'], merged_data['Longitude']))
```

```python
address = 'Toronto'

from geopy.geocoders import Nominatim 
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(merged_data['Latitude'], merged_data['Longitude'], merged_data['Borough'], merged_data['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
map_toronto
```
